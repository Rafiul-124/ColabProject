# ** Theme Ecosystem Network Analysis**

```python
from collections import defaultdict
import networkx as nx

# Create relationship network from inter_theme_relationship
relationships = defaultdict(int)

for relationship in df['inter_theme_relationship'].dropna():
    if isinstance(relationship, str):
        connections = relationship.split(', ')
        for i in range(len(connections)):
            for j in range(i+1, len(connections)):
                pair = tuple(sorted([connections[i], connections[j]]))
                relationships[pair] += 1

# Create network graph
G = nx.Graph()

for (theme1, theme2), weight in relationships.items():
    G.add_edge(theme1, theme2, weight=weight)

# Plot network
plt.figure(figsize=(16, 12))
pos = nx.spring_layout(G, k=3, iterations=50)

# Calculate node sizes based on degree
node_sizes = [G.degree(node) * 300 for node in G.nodes()]

nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='lightblue',
                      alpha=0.7, edgecolors='black')
nx.draw_networkx_edges(G, pos, width=1, alpha=0.5, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=8, font_weight='bold')

plt.title('Theme Relationship Network\n(Line thickness indicates connection strength)')
plt.axis('off')
plt.show()

# Network metrics
print("Network Analysis:")
print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
print(f"Network density: {nx.density(G):.3f}")
print("\nMost connected themes:")
for node, degree in sorted(G.degree(), key=lambda x: x[1], reverse=True)[:10]:
    print(f"  {node}: {degree} connections")
```

# **Multi-dimensional Radar Chart for Themes**

```python
from math import pi

# Select top themes for radar chart
top_themes = df.groupby('theme')['job_postings'].mean().nlargest(6).index

# Normalize metrics for radar chart
metrics = ['job_postings', 'github_stars', 'papers', 'patents', 'funding_MUSD', 'automation_risk_pct']
normalized_data = {}

for theme in top_themes:
    theme_data = df[df['theme'] == theme][metrics].mean()
    # Normalize to 0-1 scale
    normalized_data[theme] = (theme_data - df[metrics].min()) / (df[metrics].max() - df[metrics].min())

# Create radar chart
categories = metrics
N = len(categories)

angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]  # Complete the circle

fig, ax = plt.subplots(figsize=(12, 12), subplot_kw=dict(polar=True))

for theme, values in normalized_data.items():
    values = list(values)
    values += values[:1]  # Complete the circle
    ax.plot(angles, values, linewidth=2, label=theme, marker='o')
    ax.fill(angles, values, alpha=0.1)

ax.set_xticks(angles[:-1])
ax.set_xticklabels([m.replace('_', ' ').title() for m in categories])
ax.set_yticklabels([])
ax.set_title('Theme Comparison Radar Chart\n(Normalized Metrics)', size=16, y=1.1)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.show()
```

# **Advanced Correlation and Dependency Analysis**

```python
# Advanced correlation analysis with statistical significance
import scipy.stats as stats

# Calculate correlation matrix with p-values
correlation_matrix = df[numerical_cols].corr()
p_values = pd.DataFrame(index=numerical_cols, columns=numerical_cols)

for i in numerical_cols:
    for j in numerical_cols:
        if i != j:
            corr, p_value = stats.pearsonr(df[i].dropna(), df[j].dropna())
            p_values.loc[i, j] = p_value
        else:
            p_values.loc[i, j] = 0

# Create a heatmap with significance stars
plt.figure(figsize=(14, 12))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.title('Advanced Correlation Matrix\n(Upper triangle masked for clarity)')
plt.show()

# Print statistically significant correlations
print("Statistically Significant Correlations (p < 0.05):")
significant_correlations = []
for i in range(len(numerical_cols)):
    for j in range(i+1, len(numerical_cols)):
        col1, col2 = numerical_cols[i], numerical_cols[j]
        p_val = p_values.loc[col1, col2]
        if p_val < 0.05:
            corr_value = correlation_matrix.loc[col1, col2]
            significant_correlations.append((col1, col2, corr_value, p_val))

# Sort by absolute correlation strength
significant_correlations.sort(key=lambda x: abs(x[2]), reverse=True)
for col1, col2, corr, p_val in significant_correlations[:10]:
    print(f"{col1} vs {col2}: r = {corr:.3f}, p = {p_val:.4f}")
```

# Temporal Pattern Analysis

```python
# Analyze seasonal patterns and trends
from statsmodels.tsa.seasonal import seasonal_decompose

fig, axes = plt.subplots(3, 2, figsize=(16, 12))

key_metrics = ['job_postings', 'github_stars', 'funding_MUSD']
for i, metric in enumerate(key_metrics):
    if metric in monthly_data.columns:
        ts_data = monthly_data[metric].dropna()
        if len(ts_data) > 24:
            decomposition = seasonal_decompose(ts_data, period=12)

            axes[i, 0].plot(ts_data, label='Original')
            axes[i, 0].set_title(f'{metric} - Original')
            axes[i, 0].grid(True, alpha=0.3)

            axes[i, 1].plot(decomposition.trend, label='Trend', color='orange')
            axes[i, 1].plot(decomposition.seasonal, label='Seasonal', color='green')
            axes[i, 1].set_title(f'{metric} - Trend & Seasonal')
            axes[i, 1].legend()
            axes[i, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Analyze growth rates
growth_rates = monthly_data.pct_change() * 100
rolling_growth = growth_rates.rolling(window=6).mean()  # 6-month rolling average

plt.figure(figsize=(14, 8))
for metric in ['job_postings', 'github_stars', 'funding_MUSD']:
    if metric in rolling_growth.columns:
        plt.plot(rolling_growth.index, rolling_growth[metric], label=metric, linewidth=2)

plt.title('6-Month Rolling Average Growth Rates')
plt.ylabel('Growth Rate (%)')
plt.xlabel('Date')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='red', linestyle='--', alpha=0.5)
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create a mapping for world regions
world_region_mapping = {
    'United States': 'USA',
    'Europe': 'Europe',
    'Asia Pacific': 'Asia',
    'North America': 'North America',
    'Global': 'World',
    'Worldwide': 'World',
    'US early adopters': 'USA',
    'EU strong adoption': 'Europe',
    'APAC growing': 'Asia'
}

df['world_region'] = df['primary_geo'].map(world_region_mapping)

# Aggregate data for world map
world_data = df.groupby('world_region').agg({
    'job_postings': 'sum',
    'github_stars': 'mean',
    'funding_MUSD': 'sum',
    'salary_median_USD': 'mean'
}).reset_index()

# Create choropleth map for job postings
fig = px.choropleth(world_data,
                   locations='world_region',
                   locationmode='country names',
                   color='job_postings',
                   hover_name='world_region',
                   hover_data=['github_stars', 'funding_MUSD', 'salary_median_USD'],
                   color_continuous_scale='Viridis',
                   title='Global Distribution of Networking Job Postings')
fig.show()
```

`Advanced Geographical Heatmaps`

```python
# Create a comprehensive geographical dashboard
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('Job Postings by Region', 'Total Funding by Region',
                   'Average Salary by Region', 'GitHub Stars by Region'),
    specs=[[{"type": "choropleth"}, {"type": "choropleth"}],
           [{"type": "choropleth"}, {"type": "choropleth"}]]
)

# Job Postings
fig.add_trace(go.Choropleth(
    locations=world_data['world_region'],
    z=world_data['job_postings'],
    locationmode='country names',
    colorscale='Blues',
    colorbar_title="Job Postings",
    showscale=False
), row=1, col=1)

# Total Funding
fig.add_trace(go.Choropleth(
    locations=world_data['world_region'],
    z=world_data['funding_MUSD'],
    locationmode='country names',
    colorscale='Greens',
    colorbar_title="Funding (M USD)",
    showscale=False
), row=1, col=2)

# Average Salary
fig.add_trace(go.Choropleth(
    locations=world_data['world_region'],
    z=world_data['salary_median_USD'],
    locationmode='country names',
    colorscale='Reds',
    colorbar_title="Salary (USD)",
    showscale=False
), row=2, col=1)

# GitHub Stars
fig.add_trace(go.Choropleth(
    locations=world_data['world_region'],
    z=world_data['github_stars'],
    locationmode='country names',
    colorscale='Purples',
    colorbar_title="GitHub Stars",
    showscale=False
), row=2, col=2)

fig.update_layout(height=800, title_text="Global Networking Trends Analysis")
fig.show()
```
